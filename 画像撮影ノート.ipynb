```python
lm = input("刃物長さを入れてください mm = ")
ls = input("刃物の右端までの距離を入れてください mm = ")


R = 200 #1回転に必要なパルス数
lsp = int(ls) * int(R) #刃物右端までのパルス数

# ここでラメラを刃物右はまで移動する

lp = 3200 #カメラを20ｍｍ動かす時のパルス数（（移動量20mm　/　ネジピッチ　1.25ｍｍ）*　1回転パルス数　200））

lmp = int(lm) * int(R) #刃物長さのパルス数
lmax = lmp + lsp

i = 0
while i < lmax:

   # ここで写真を撮って保存する

   # 引続きカメラを"lp"移動する



    i = i + lp

    print("移動距離")
    print(i)
    print("現在位置")
    print(i + lsp)

    if i == lmax:
      break  # ループを終了

print(ls)
print(lsp)
print(lm)
print(lmp)
```
